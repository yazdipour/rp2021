```
def findResult(answer):
  try: 
    if 'boolean' in str(answer): return answer['boolean']
    else: return answer['results']['bindings'][0][answer['head']['vars'][0]]['value']
  except: 
    print('Failure - Returning None:' + str(answer));return None
```

```
findResult(df.iloc[99].answers[0])
```

```
findResult(df[df.apply(lambda x: '{}' in str(x.answers[0]['head']), axis=1)].iloc[0].answers[0])
```

```
x= df.answers.apply(lambda x: findResult(x[0]))
```
